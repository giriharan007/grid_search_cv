```python
#import the dependencies
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

```python
breast_cancer_dataset=sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
#loading the data to data frame
data_frame=pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
#adding the 'target' column to the data frame
data_frame['label']=breast_cancer_dataset.target
```

```python
data_frame.head()
```

```python
#check the shape of the dataframe that is rows and columns
data_frame.shape
```

```python
#checking there is any missing values 
data_frame.isnull().sum()
```

```python
#checking the distribution of target values
data_frame['label'].value_counts()
```

```python
#load the features and label in x and y
x=data_frame.drop(columns='label',axis=1)
y=data_frame['label']
```

```python
print(x)
```

```python
print(y)
```

```python
x=np.asarray(x)
y=np.asarray(y)
```

```python
#load svc]
model=SVC()
```

```python
#create dictionary with hyperparameter values
parameters={
    "kernel":["linear", "poly","rbf","sigmoid"],
    "C":[1,5,10,20] 
    
}  
```

```python
classifier=GridSearchCV(model,parameters,cv=5)
```

```python
#fitting the data to our model
classifier.fit(x,y)
```

```python
classifier.cv_results_
```
